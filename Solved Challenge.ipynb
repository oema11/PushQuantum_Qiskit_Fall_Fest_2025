## 2. Single-Qubit Gates and Phases

**Explanation:** Single-qubit gates like X, Y, Z, H, S, and T are basic operations on one qubit. S and T are phase gates. The S-gate adds a π/2 phase to the |1⟩ component of any quantum state, while the T-gate adds a π/4 phase to the |1⟩ component, leaving the |0⟩ component unchanged in both cases. These phase shifts are crucial for many quantum algorithms.

## 3. Superposition and Bloch Sphere Rotations

A rotation by an angle θ around the Y-axis (`RY(θ)`) on an initial state |0⟩ produces the superposition cos(θ/2)|0⟩ + sin(θ/2)|1⟩. The probabilities of measuring 0 or 1 are the squares of these amplitudes.

### Grover's algorithm applied to the problem of finding perfect numbers ###

For this challenge we apply Grovers algorithm to find what is known as perfect numbers:
A perfect number is defined as a number, that is the sum of its divisors, e.g.:
$$
divisor(6)=\{3,2,1\}\quad  and \quad 6=3+2+1
$$
Therefore 6 is a perfect number.

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.circuit.library import grover_operator
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_distribution
from qiskit_ibm_runtime import Sampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

def perfect_numbers(max_value): 
    perfects = []
    width = len(format(max_value - 1, 'b'))
    for num in range(2, max_value):
        divisors = [i for i in range(1, num) if num % i == 0]
        if sum(divisors) == num:
            perfects.append(format(num, f'0{width}b'))
    return perfects

def grover_oracle(marked_states):
    if not isinstance(marked_states, list):
        marked_states = [marked_states]
    num_qubits = len(marked_states[0])
    qc = QuantumCircuit(num_qubits)
    for target in marked_states:
        rev_target = target[::-1]
        zero_inds = [i for i, bit in enumerate(rev_target) if bit == "0"]
        qc.x(zero_inds)
        qc.h(num_qubits - 1)
        qc.mcx(list(range(num_qubits - 1)), num_qubits - 1)
        qc.h(num_qubits - 1)
        qc.x(zero_inds)
    return qc

# parameters
max_value = 16  # search space up to 15
marked_states = perfect_numbers(max_value)
marked_states = [s[::-1] for s in marked_states]  # fix Qiskit bit order

print("Marked (perfect) states:", marked_states)

N = int(np.ceil(np.log2(max_value)))
qc = QuantumCircuit(N, N)
qc.h(range(N))

oracle = grover_oracle(marked_states)
grover_op = grover_operator(oracle)
iterations = 3
for _ in range(iterations):
    qc.compose(grover_op, inplace=True)
qc.measure_all()

# simulate
backend = AerSimulator()
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)
sampler = Sampler(mode=backend)
job = sampler.run([isa_circuit], shots=2048)
result = job.result()
counts = result[0].data.meas.get_counts()
counts_reversed = {k[::-1]: v for k, v in counts.items()}

# plot results
plot_distribution(counts_reversed)
```

In the case of an ideal AerSimulator we get a clear result for 0110, which is 6. Let us try higher numbers:

```python
max_value = 513 # search space up to 512
marked_states = perfect_numbers(max_value)
marked_states = [s[::-1] for s in marked_states]  # fix Qiskit bit order

print("Marked (perfect) states:", marked_states)

N = int(np.ceil(np.log2(max_value)))
qc = QuantumCircuit(N, N)
qc.h(range(N))

oracle = grover_oracle(marked_states)
grover_op = grover_operator(oracle)
iterations = 3
for _ in range(iterations):
    qc.compose(grover_op, inplace=True)
qc.measure_all()


# simulate
backend = AerSimulator()
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)
sampler = Sampler(mode=backend)
job = sampler.run([isa_circuit], shots=2048)
result = job.result()
counts = result[0].data.meas.get_counts()
counts_reversed = {k[::-1]: v for k, v in counts.items()}

# plot results
plot_distribution(counts_reversed, number_to_keep=20)
```

We again get $$000000110_2=6_{10},$$ but also $$000011100_2=28_{10}$$ and $$111110000_2=496_{10}$$. From literature we can see that those are indeed perfect numbers as well. Due to the big sample size, most of the remaining samples were fitted to 'rest'. Their relative probabilities of course stack up, but there is no other significant number in there, that might be perfect. Let's look at a more realistic example:

```python
max_value = 16 # search space up to 15
marked_states = perfect_numbers(max_value)
marked_states = [s[::-1] for s in marked_states]  # fix Qiskit bit order

print("Marked (perfect) states:", marked_states)

N = int(np.ceil(np.log2(max_value)))
qc = QuantumCircuit(N, N)
qc.h(range(N))

oracle = grover_oracle(marked_states)
grover_op = grover_operator(oracle)
iterations = 3
for _ in range(iterations):
    qc.compose(grover_op, inplace=True)
qc.measure_all()


# simulate
backend = AerSimulator.from_backend(FakeBrisbane())
pm = generate_preset_pass_manager(backend=backend, optimization_level=3)
isa_circuit = pm.run(qc)
sampler = Sampler(mode=backend)
job = sampler.run([isa_circuit], shots=2048)
result = job.result()
counts = result[0].data.meas.get_counts()

# plot results
plot_distribution(counts, number_to_keep=20)
```

We get good results, but at the same time, we notice higher fluctuations. An increase of the iterations did not yield better results. Increasing the number of samples, quickly gave rise to randomness and the results were useless.
